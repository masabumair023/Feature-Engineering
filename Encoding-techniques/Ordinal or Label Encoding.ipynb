```python
import pandas as pd
import datetime
```

```python
# Create a list of dates with 20 days differnece from today and then
# transform it into a dataframe

df_base = datetime.datetime.today()
df_date_list = [df_base - datetime.timedelta(days = x) for x in range(0,20)]
df = pd.DataFrame(df_date_list)
df.columns = ['Day']
```

```python
df
```

```python
# Extract week day names
df['week day'] = df['Day'].dt.strftime("%A")
```

```python
df
```

```python
# Applying Label Encoding technique

weekday_map = {
    'Monday' : 1,
    'Tuesday' : 2,
    'Wednesday' : 3,
    'Thursday' : 4,
    'Friday' : 5,
    'Saturday' : 6,
    'Sunday' : 7,
}

df["weekday ordinal"] = df['week day'].map(weekday_map)
```
